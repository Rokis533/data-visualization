```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Data.csv", sep=';')
```

```python
df = pd.DataFrame({'x': data.ViewedBy,'y': data.RememberedBy})
plt.style.use('_mpl-gallery-nogrid')

x = df.x
y = df.y

# plot:
fig, ax = plt.subplots()
ax.set_ylabel('Išsaugojimai')
ax.set_xlabel('Peržiūros')
ax.set_title('Koreliacija tarp išsaugojimu ir peržiūrų', pad=20)
ax.hist2d(x, y, bins=(np.arange(0, 3000, 100), np.arange(0, 100, 4)))

ax.set(xlim=(0, 3000), ylim=(0, 100))
fig.set_size_inches(5.5, 5.5, forward=True)
plt.show()
```

```python
df = pd.DataFrame({'Price': data.Price,'Date': data.Year}).sort_values(by=['Date'])
plt.style.use('_mpl-gallery')

# make data
x = df.Date
y = df.Price

# plot
fig, ax = plt.subplots()
fig.set_dpi(100)
ax.set_ylabel('Kaina')
ax.set_xlabel('Statybos metai')
ax.set_title('Koreliacija tarp kainos ir namo statybos metų', pad=20)
fig.set_size_inches(10.5, 3.5, forward=True)
ax.plot(x, y, linewidth=2.0)



plt.show()
```

```python
df = pd.DataFrame({'Area': data.Area,'Average prices': data.Price}).sort_values(by=['Average prices'])
df.boxplot(by ='Area', column =['Average prices'], rot=90, figsize=(16,8))
mean_df = df.groupby('Area').mean().reset_index()
fig.set_size_inches(15.5, 6.5, forward=True)
plt.title("")
plt.show()
```

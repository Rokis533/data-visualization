```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

area = []
price = []
data= pd.read_csv("Data.csv", sep=';')
```

```python
df = pd.DataFrame({'Area': data.Area,'Average prices': data.Price}).sort_values(by=['Average prices'])
df.boxplot(by ='Area', column =['Average prices'], rot=90)
mean_df = df.groupby('Area').mean().reset_index()

plt.title("")
plt.show()
```
